```python
import numpy as np
import pandas as pd
import libcirctis

tis_transformer_file_result = 'TIS_Transformer/results.npy'
test_dataset_samples_file = 'datasets/splits/test/samples.tsv'

output_samples_eval_file_atg = 'outputs/circTIS_samples_eval_ATG.tsv'
output_samples_eval_file_nc = 'outputs/circTIS_samples_eval_NC.tsv'
```

```python
tt_results =  np.load(tis_transformer_file_result, allow_pickle=True)
print(tt_results)
```

```python
ls = tt_results.tolist()
df = pd.DataFrame(ls)
df
```

```python
df.columns = ['score', 'one', 'id']
df
```

```python
df = df.drop('one', axis=1)
df
```

```python
df2 = df.explode('id')
df2
```

```python
df3 = df2.explode('score')
df3
```

```python
df4 = df3.explode('score')
df4
```

```python
ls_pos = []
pos = 0
previous_id = ''

for index, row in df4.iterrows():

    if row['id'] != previous_id:
        pos = 1
    else:
        pos += 1
    
    ls_pos.append(pos)
    previous_id = row['id']
```

```python
len(ls_pos)
```

```python
df4['position'] = ls_pos
df4
```

```python
df4.reset_index(drop=True, inplace=True)
df4
```

```python
df_samples = pd.read_csv(test_dataset_samples_file, sep='\t', header=0)
df_samples
```

```python
df_samples_merge = pd.merge(df_samples, df4, left_on=['circrna_id', 'TIS_position'], right_on=['id', 'position'])
df_samples_merge
```

```python
df_samples_eval = df_samples_merge[['circrna_id', 'TIS_type', 'TIS_position', 'sample_label', 'score']].copy()
df_samples_eval.rename(columns={'score': 'TIS_Transformer_score'}, inplace=True)
df_samples_eval
```

```python
df_samples_eval_atg = df_samples_eval[df_samples_eval['TIS_type'] == 'ATG']
df_samples_eval_atg.to_csv(output_samples_eval_file_atg, sep='\t', index=False)
df_samples_eval_atg.shape
```

```python
metrics = {}
libcirctis.calc_metrics_scores(metrics, df_samples_eval_atg['sample_label'], df_samples_eval_atg['TIS_Transformer_score'])
metrics
```

```python
df_samples_eval_nc = df_samples_eval[df_samples_eval['TIS_type'] != 'ATG']
df_samples_eval_nc.to_csv(output_samples_eval_file_nc, sep='\t', index=False)
df_samples_eval_nc.shape
```

```python
metrics = {}
libcirctis.calc_metrics_scores(metrics, df_samples_eval_nc['sample_label'], df_samples_eval_nc['TIS_Transformer_score'])
metrics
```
