```python
import pandas as pd
from matplotlib import pyplot
import seaborn as sns


result_file = 'outputs/08.tsv'
parameter_varying = 'down_size'

pd.set_option('display.max_columns', None)
# pd.set_option('display.max_rows', None)
```

```python
df_eval = pd.read_csv(result_file, sep='\t')
```

```python
# CV means ordered
df_eval_cv = df_eval.groupby(parameter_varying).mean()
df_eval_cv = df_eval_cv.sort_values('F1-score', ascending=False)
df_eval_cv.head(30)
```

```python
# Line chart

dims = (10, 10)

x_axis = parameter_varying
y_axis = ['F1-score', 'Precision', 'Recall']


fig, ax = pyplot.subplots(figsize=dims)
pyplot.grid()

for col in y_axis:
    sns.lineplot(data=df_eval, x=x_axis, y=col, legend='brief',label=col, marker="o")

ax.set_ylabel('')
ax.set_xlabel('Downstream sample length')
```
